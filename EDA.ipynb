```python
ratings_with_books = ratings.merge(books,on = 'ISBN')
```

```python
num_ratings = ratings_with_books.groupby('Title')['Rating'].count().reset_index()
```

```python
num_ratings
```

```python
num_ratings.rename(columns={"Rating":"Num_of_rating"},inplace=True)
```

```python
num_ratings
```

```python
Final_Rating = ratings_with_books.merge(num_ratings,on = 'Title')
```

```python
Final_Rating
```

```python
Final_Rating.shape
```

```python
Final_Rating = Final_Rating[Final_Rating['Num_of_rating']>=50]
```

```python
Final_Rating.sample()
```

```python
Final_Rating.drop_duplicates(['Title','User_ID'],inplace=True)
```

```python
Final_Rating.shape
```

```python
book_pivot = Final_Rating.pivot_table(columns='User_ID',index='Title',values='Rating')
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm = 'brute')
```

```python
model.fit(book_sparse)
```

```python
distance,suggetion = model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1),n_neighbors=6)
```
